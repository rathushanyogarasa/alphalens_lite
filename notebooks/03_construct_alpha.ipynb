# 03 Construct Alpha
Build technical + sentiment factors and fuse into a single `alpha_score`.

```python
import pandas as pd
import config
from src.market_data import run_market_data
from src.technical_factors import run_technical_factors
from src.factor_engine import run_factor_engine

print('processed dir:', config.PROCESSED_DIR)
```

## 1) Build market panel
`refresh=False` uses cached prices when available for faster runs.

```python
prices, market_panel = run_market_data(
    tickers=config.TICKERS,
    start_date=config.MARKET_START_DATE,
    end_date=config.MARKET_END_DATE,
    refresh=False,
    save=True,
)
print(prices.shape, market_panel.shape)
market_panel.head()
```

## 2) Build technical factors

```python
technical = run_technical_factors(prices=prices, save=True)
print(technical.shape)
technical.head()
```

## 3) Load sentiment factor and fuse all factors
Run notebook 02 first so `sentiment_factor.csv` exists.

```python
sentiment = pd.read_csv(config.PROCESSED_DIR / 'sentiment_factor.csv')
alpha = run_factor_engine(technical_df=technical, sentiment_df=sentiment, save=True)
print(alpha.shape)
alpha.head()
```

```python
alpha['date'] = pd.to_datetime(alpha['date'])
alpha.groupby('date')['alpha_score'].agg(['mean','std']).tail()
```
