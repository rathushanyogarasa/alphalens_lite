# 04 Backtest
Run fast long/short backtest on prebuilt alpha factors.

```python
import pandas as pd
import config
from src.backtest import run_backtest

alpha = pd.read_csv(config.PROCESSED_DIR / 'alpha_factors.csv')
market_panel = pd.read_csv(config.PROCESSED_DIR / 'market_panel.csv')

daily, positions = run_backtest(
    alpha_df=alpha,
    market_panel=market_panel,
    hold_days=config.HOLD_DAYS,
    quantile_cutoff=config.QUANTILE_CUTOFF,
    transaction_cost_bps=config.TRANSACTION_COST_BPS,
    slippage_bps=config.SLIPPAGE_BPS,
    save=True,
)
print(daily.shape, positions.shape)
daily.tail()
```
